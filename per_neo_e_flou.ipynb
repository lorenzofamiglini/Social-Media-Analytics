```python
import pandas as pd
import numpy as np
import re
import networkx as nx
import matplotlib.pyplot as plt
import community
from collections import Counter
```

```python
df_for_men=pd.read_csv("cluster_mention.csv")
```

```python
df_for_hash=pd.read_csv("cluster_hashtag.csv")
```

```python
df_for_men.columns
```

```python
df_for_hash.columns
```

```python
df_for_hash["hashtags"] = df_for_hash["hashtags"].replace(" ","", regex = True).str.lower()
df_for_hash["hashtags"] = df_for_hash["hashtags"].replace("\[","", regex = True).str.lower()
df_for_hash["hashtags"] = df_for_hash["hashtags"].replace("\]","", regex = True).str.lower()
df_for_hash["hashtags"] = df_for_hash["hashtags"].replace("\'","", regex = True).str.lower()
df_for_hash["hashtags"] = df_for_hash["hashtags"].apply(lambda x: x.split(','))
```

```python
df_for_men["@mentions"] = df_for_men["@mentions"].replace(" ","", regex = True).str.lower()
df_for_men["@mentions"] = df_for_men["@mentions"].replace("\[","", regex = True).str.lower()
df_for_men["@mentions"] = df_for_men["@mentions"].replace("\]","", regex = True).str.lower()
df_for_men["@mentions"] = df_for_men["@mentions"].replace("\'","", regex = True).str.lower()
df_for_men["@mentions"] = df_for_men["@mentions"].apply(lambda x: x.split(','))
```

```python
GH = nx.Graph()
for r in df_for_hash.iterrows():
    for user in r[1]['hashtags']:
        GH.add_edge(r[1]['screen_name'], user)
```

```python
GM = nx.Graph()
for r in df_for_men.iterrows():
    for user in r[1]['@mentions']:
        GM.add_edge(r[1]['screen_name'], user)
```

```python
nodesh = pd.DataFrame( np.concatenate( (edgeh["from"], edgeh["to"]), axis=0 ) )
nodesh.columns = [ 'id' ]
nodesh=pd.DataFrame(nodesh["id"].unique())
nodesh.columns = [ 'id' ]
nodesh.head()
```

```python
result=nodesh.merge(df_for_hash,left_on="id", right_on="screen_name",how="left")[["id","cluster"]]
```

```python
result["cluster"] = result["cluster"].fillna(1000)
```

```python
result.drop_duplicates(inplace=True)
```

```python
result["cluster"] = result["cluster"].astype(int)
```

```python
result
```

```python
result= result[(result["cluster"]<4) | (result["cluster"]==10) | (result["cluster"]==11)| (result["cluster"]==1000)]
```
