```python
#eliminazione mentions dal testo (@...), url e emoji (<..>)
def remove_mentions(x):
    p = re.sub(r'@([^\s:]+)','', x)
    return p

def remove_url(x):
    p = re.sub(r'http\S+','', x)
    return p

def remove_emoji(x):
    p = re.sub(r'(\<u+\S*\>)', '',x)
    return p tweet["text_def"] = tweet["text"].apply(remove_mentions)
```
